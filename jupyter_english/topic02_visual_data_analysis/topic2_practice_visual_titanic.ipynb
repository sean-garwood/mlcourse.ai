```python
import numpy as np
import pandas as pd
import seaborn as sns

sns.set()
import matplotlib.pyplot as plt
```

**Read data**

```python
train_df = pd.read_csv("../../data/titanic_train.csv", index_col="PassengerId")
```

```python
train_df.head(2)
```

```python
train_df.describe(include="all")
```

```python
train_df.info()
```

**Let's drop`Cabin`, and then – all rows with missing values.**

```python
train_df = train_df.drop("Cabin", axis=1).dropna()
```

```python
train_df.shape
```

**1. Build a picture to visualize all scatter plots for each pair of features `Age`, `Fare`, `SibSp`, `Parch` and `Survived`. ( `scatter_matrix ` from Pandas or `pairplot` from Seaborn)**

```python
features1 = ['Age', 'Fare', 'SibSp', 'Parch', 'Survived']
print(features1)
%config InlineBackend.figure_format = 'png'
sns.pairplot(train_df[features1], hue="Survived")
```

**2. How does ticket price (`Fare`) depend on `Pclass`? Build a boxplot.**

```python
sns.boxplot(y="Fare", x="Pclass", data=train_df)

# the ticket price is directly correlated with the class of the ticket. The
# higher the class, the higher the ticket price.
```

**3. Let's build the same plot but restricting values of `Fare` to be less than 95% quantile of the initial vector (to drop outliers that make the plot less clear).**

```python
# get the 95th percentile of Fare
fare_95 = train_df["Fare"].quantile(0.95)
train_df_fare_95 = train_df[train_df["Fare"] < fare_95]
sns.boxplot(x='Pclass', y='Fare', data=train_df_fare_95)
```

**4. How is the percentage of surviving passengers dependent on passengers' gender? Depict it with `Seaborn.countplot` using the `hue` argument.**

```python
sns.countplot(x='Sex', data=train_df, hue='Survived')

# The proportion of surviving females to all females is much higher than it is
# for men. This is a strong indicator that there is a correlation between 'Sex'
# and 'Survived'.
```

**5. How does the distribution of ticket prices differ for those who survived and those who didn't. Depict it with `Seaborn.boxplot`**

```python
sns.boxplot(x='Survived', y='Fare', data=train_df_fare_95)
```

**6. How does survival depend on passengers' age?  Verify (graphically) an assumption that youngsters (< 30 y.o.) survived more frequently than old people (> 55 y.o.).**

```python
old = train_df[train_df["Age"] > 55]
youngsters = train_df[train_df["Age"] < 30]
# plot age versus survival
_, ax = plt.subplots(1, 2, figsize=(12, 6))
sns.countplot(x='Survived', data=old, ax=ax[0])
ax[0].set_title("Old people")
sns.countplot(x='Survived', data=youngsters, ax=ax[1])
ax[1].set_title("Young people")
```
